This notebook demonstrates a network vulnerability scanner built using Python and Nmap. 
It scans a target system and identifies open ports and services.

```python
import nmap

scanner = nmap.PortScanner()
target = "192.168.56.101"

print("Scanning target:", target)
scanner.scan(target, "1-100")

for host in scanner.all_hosts():
    print("\nHost:", host)
    for port in scanner[host]['tcp']:
        print("Port:", port, "| State:", scanner[host]['tcp'][port]['state'])
```

```python
import nmap

scanner = nmap.PortScanner()
target = "192.168.56.101"

print("Scanning target:", target)

scanner.scan(target, "1-100")

for host in scanner.all_hosts():
    print("\nHost:", host)
    
    if 'tcp' in scanner[host]:
        for port in scanner[host]['tcp']:
            state = scanner[host]['tcp'][port]['state']
            print("Port:", port, "| State:", state)
```

The output shows open ports such as FTP (21), SSH (22), and HTTP (80), 
which indicate services running on the target system.
